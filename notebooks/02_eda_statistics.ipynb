# 02 - EDA and Statistics

This notebook demonstrates descriptive statistics, distribution shape, outlier behavior, correlations, and fraud vs non-fraud statistical comparisons.

```python
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from src.statistical_analysis import StatisticalAnalyzer

df = pd.read_csv(Path("../data/raw/creditcard.csv"))
analyzer = StatisticalAnalyzer()
```

```python
descriptive_df = analyzer.descriptive_statistics(df)
descriptive_df.head(12)
```

```python
skew_kurtosis = descriptive_df[['feature', 'skewness', 'kurtosis']].sort_values('kurtosis', ascending=False)
skew_kurtosis.head(10)
```

```python
outlier_df = analyzer.outlier_summary_iqr(df)
outlier_df.head(10)
```

```python
correlation_df = analyzer.correlation_analysis(df)
correlation_df.head(10)
```

```python
comparison_df = analyzer.fraud_nonfraud_comparison(df)
comparison_df
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=correlation_df.head(12), x='feature', y='correlation_with_class', palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Top Features by Correlation with Fraud Class')
plt.show()
```

## Statistical Interpretation

- Features with low p-values in Mann-Whitney U tests suggest distributional differences between fraud and non-fraud classes.
- IQR outlier summary highlights where extreme transaction behavior is concentrated.
- Precision and recall remain more useful than accuracy for this problem due to severe class imbalance.
